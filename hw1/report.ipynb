## Question 1

Let $i_k$ be the slew rate at time $t_k$. Then, the equations of linear time-invariant system are given as, 
<br>
$x_{k+1} = A_k x_k + B_k u_k$
<br>
$u_{k+1} = u_i + \Delta t i_k$

Let define new variables as 
<br>
$\bar{x}_k = 
\begin{cases}
\left[ x_k, u_k \right]^{\top} & \text{for } k<N\\
\left[ x_k, 0 \right]^{\top} & \text{for } k=N
\end{cases}
$
<br>
$\bar{u}_k = \left[0, i_k\right]^{\top}$. 
<br>
Then, the above equations are formulated as,
<br>
$\bar{x}_{k+1} = \bar{A}_k \bar{x}_k + \bar{B}_k \bar{u}_k$,
<br>
where
<br>
$\bar{A}_k = 
\begin{bmatrix}
A_k & B_k \\
O & I 
\end{bmatrix}
$,
$\bar{B}_k 
=
\Delta t
\begin{bmatrix}
O & O\\
O & I
\end{bmatrix}
$.

From the above formulation, we can apply a quadratic cost with respect to $\bar{x}_k$ and $\bar{u}_k$. To regulate $i_k$, $u_k$ and $x_k$ to the zero point, the quadratic cost can be given as,
<br>
$
c(\bar{x}_k, \bar{u}_k) 
=
\begin{cases}
{1\over2} \bar{x}_k^\top Q_k \bar{x}_k + {1\over2} \bar{u}_k^\top R_k \bar{u}_k  & k<N \text{ (stage-wise cost)}\\
{1\over2} \bar{x}_k^\top P_N \bar{x}_k  & k=N \text{ (terminal cost)}
\end{cases}
$. 
<br>
Here, we consider $Q_k$, $R_k$ and $P_N$ to be positive definite. 
<br>

By plugging these results into the LQR formulation of an LTI system, the optimal sequence of $\{\bar{u}_{k}\}_{k=1}^N$ is given as, 
<br>
$
\bar{u}_{k}^{*} = L_k \bar{x}_k
$ for $k\leq N-1$,
<br>
where
<br>
$L_k = -\left( R_k + \bar{B}_k^\top K_{k+1} \bar{B}_k \right)^{-1} \left( \bar{B}_k^\top K_{k+1} \bar{B}_k \right)$ for $k \leq N-1 $,
<br>
$K_k = 
\begin{cases} 
Q_k + L_k^\top R_k L_k + \left( \bar{A}_k + \bar{B}_k L_k \right) ^\top K_{k+1} \left(\bar{A}_k + \bar{B}_k L_k \right) & \text{for } k \leq N-1 \\
P_N & \text{for } k =N
\end{cases}
$.
<br>
$\bar{u}_{k}^{*}$ can be computed backward from $k=N-1$ to $k=1$. 
<br>
Then, the optimal control sequence $u_{k}^{*}$ can be computed forward from $k=1$ to $k=N-1$, by the equation
<br>
$\bar{x}_{k+1} = \bar{A}_k \bar{x}_k + \bar{B}_k \bar{u}_k$.



## Question 2

Define a new variable,as
<br>
$\bar{x}_k = 
\begin{cases}
\left[ x_k, u_k \right]^{\top} & \text{for } k<N\\
\left[ x_k, 0 \right]^{\top} & \text{for } k=N
\end{cases}
$.
<br>
Then, we can re-write the formulations of the given dynamics and the quadratic cost with respect $\bar{x}_k$, as
<br>
$
\bar{x}_{k+1} = \bar{A}_k \bar{x}_k + \bar{d}_k
$,
<br>
$
c(\bar{x}_k) 
=
{1\over2} 
\bar{x}_k^\top
\bar{Q}_k
\bar{x}_k
+
\bar{q}_k^\top
\bar{x}_k
+
h_k
$,
<br>
where 
<br>
$
\bar{A}_k = 
\begin{bmatrix}
A_k & B_k \\
O & O
\end{bmatrix}
$,
$
\bar{Q}_k = 
\begin{bmatrix}
Q_k & H \\
H^\top & R_k
\end{bmatrix}
$, 
$
\bar{d}_k = 
\begin{bmatrix}
d_k \\
0
\end{bmatrix}
$,
$
\bar{q}_k = 
\begin{bmatrix}
q_k \\
r_k
\end{bmatrix}
$.
<br>

Let $J_k$ be the cost-to-go at $k$-th step. Then, $J_{N-1}$ is given as,
<br>
$
J_{N-1} = {1\over2} \bar{x}_{N-1}^\top \bar{Q}_{N-1} \bar{x}_{N-1} + \bar{q}_{N-1}^\top + h_{N-1} + J_{N}
$
<br>
By using the equation $\bar{x}_{k+1} = \bar{A}_k \bar{x}_k + \bar{d}_k$, we can get 
<br>
$
\begin{align*}
J_{N-1}
&= {1\over2} \bar{x}_{N-1}^\top \left[ \bar{Q}_{N-1} + \bar{A}_{N-1}^\top \bar{Q}_{N} \bar{A}_{N-1}\right]\bar{x}_{N-1}
+ \left[\bar{q}_{N-1}^\top + \bar{d}_{N-1}^\top \bar{Q}_{N} \bar{A}_{N-1} + \bar{q}_{N} \bar{A}_{N-1}  \right] \bar{x}_{N-1}
+ h_{N-1} + h_{N} - \bar{q}_{N} \bar{d}_{N-1} + {1 \over 2} \bar{d}_{N-1}^\top \bar{Q}_{N} \bar{d}_{N-1}\\
& = {1\over2} \bar{x}_{N-1}^\top V_{N-1} \bar{x}_{N-1}+ v_{N-1}^\top \bar{x}_{N-1} + w_{N-1}   
\end{align*}
$, 
<br>
where
<br>
$
V_N = \bar{Q}_{N-1} + \bar{A}_{N-1}^\top \bar{Q}_{N} \bar{A}_{N-1}
$,
<br>
$
v_N = \left[\bar{q}_{N-1}^\top + \bar{d}_{N-1}^\top \bar{Q}_{N} \bar{A}_{N-1} + \bar{q}_{N} \bar{A}_{N-1}  \right]^\top
$,
<br>
$
w_N = h_{N-1} + h_{N} - \bar{q}_{N} \bar{d}_{N-1} + {1 \over 2} \bar{d}_{N-1}^\top \bar{Q}_{N} \bar{d}_{N-1}
$.

Likewise, by computing backward from $k=N-1$, $J_{k}$ is given as,
<br>
$
\begin{align*}
J_{k} & = {1\over2} \bar{x}_{k}^\top \bar{Q}_{k} \bar{x}_{k} + \bar{q}_{k}^\top + h_{k} + J_{k+1}\\
&= {1\over2} \bar{x}_{k}^\top \left[ \bar{Q}_{k} + \bar{A}_{k}^\top \bar{Q}_{k+1} \bar{A}_{k}\right]\bar{x}_{k}
+ \left[\bar{q}_{k}^\top + \bar{d}_{k}^\top \bar{Q}_{k+1} \bar{A}_{k} + \bar{q}_{k+1} \bar{A}_{k}  \right] \bar{x}_{k}
+ h_{k} + h_{k+1} - \bar{q}_{k+1} \bar{d}_{k} + {1 \over 2} \bar{d}_{k}^\top \bar{Q}_{k+1} d_{k}\\
& = {1\over2} \bar{x}_{k}^\top V_{k} \bar{x}_{k}+ v_{k}^\top \bar{x}_{k} + w_{k}   
\end{align*}
$,
<br>
where
<br>
$
V_k = \bar{Q}_{k} + \bar{A}_k^\top \bar{Q}_{k+1} \bar{A}_k
$,
<br>
$
v_k = \left[\bar{q}_{k}^\top + \bar{d}_{k}^\top \bar{Q}_{k+1} \bar{A}_{k} + \bar{q}_{k+1} \bar{A}_{k}  \right]^\top
$,
<br>
$
w_k = h_{k} + h_{k+1} - \bar{q}_{k+1} \bar{d}_{k} + {1 \over 2} d_{k}^\top \bar{Q}_{k+1} \bar{d}_{k}
$.

Thus, for all $k \leq N$, the above cost-to-go can be written as,
<br>
$
c(\bar{x}_k) 
=
{1\over2} 
(\bar{x}_k -\bar{b}_k)^\top
V_k
(\bar{x}_k -\bar{b}_k)
+
w_k - {1 \over 2} \bar{b}_k^\top V_k \bar{b}_k
$.
<br>
where
<br>
$\bar{b}_k = -\left(V_k + V_k^\top \right) ^{-1} v^\top $,
<br>
which is minimized at $\bar{x}_k = \bar{b}_k$.

From the above results, the optimal sequence of $\{\bar{x}_k\}_{k=1}^{N}$ can be computed backward from $k=N$, as
<br>
$\bar{x}_k^{*} = - \left(V_k + V_k^\top \right) ^{-1} v^\top$,
$
V_k = \bar{Q}_{k} + \bar{A}_{k}^\top \bar{Q}_{k+1} \bar{A}_{k}
$,
<br>
$
v_k = \left[\bar{q}_{k}^\top + \bar{d}_{k}^\top \bar{Q}_{k+1} \bar{A}_{k} + \bar{q}_{k+1} \bar{A}_{k}  \right]^\top
$,
<br>
$
w_k = h_{k} + h_{k+1} - \bar{q}_{k+1} \bar{d}_{k} +{1 \over 2} \bar{d}_{k}^\top \bar{Q}_{k+1} \bar{d}_{k}
$.
<br>
Then, the optimal control sequence $u_{k}^{*}$ can be found at $\{\bar{x}_k^{*}\}_{k=1}^{N-1}$.

## Question 3

The differential equation of the dynamics are given as
<br>
$
\begin{bmatrix} 
    F \\ 0\\
  \end{bmatrix} = 
  \begin{bmatrix}
    M_t & -ml \cos\theta \\
    -ml \cos\theta & J_t\\
  \end{bmatrix}
  \begin{bmatrix}
    \ddot{q} \\
    \ddot{\theta}\\
  \end{bmatrix}
  +
  \begin{bmatrix}
    c & 0 \\
    0 & \gamma\\
  \end{bmatrix}
  \begin{bmatrix}
    \dot{q} \\
    \dot{\theta}\\
  \end{bmatrix}
  +
  \begin{bmatrix}
    ml\dot{\theta}^2\sin{\theta} \\
    -mgl \sin{\theta}\\
  \end{bmatrix}
$.
<br>

We can linearlize the equation as,
<br>
$
\begin{align*}
\begin{bmatrix} 
    F \\ 0\\
\end{bmatrix} 
&\approx 
\begin{bmatrix}
    M_t & -ml \cos\theta \\
    -ml \cos\theta & J_t\\
\end{bmatrix}
\begin{bmatrix}
    \ddot{q} \\
    \ddot{\theta}\\
\end{bmatrix}
+
\begin{bmatrix}
    c & 0 \\
    0 & \gamma\\
\end{bmatrix}
\begin{bmatrix}
    \dot{q} \\
    \dot{\theta}\\
\end{bmatrix}
+
\dot{\theta}
\begin{bmatrix}
    0 \\
    -mgl \cos{\theta}\\
\end{bmatrix}
\\
&= 
C\left(q, \theta\right) 
\begin{bmatrix}
    \ddot{q} \\
    \ddot{\theta}\\
\end{bmatrix}
+
D\left(\theta\right) 
\begin{bmatrix}
    \dot{q} \\
    \dot{\theta}\\
\end{bmatrix}
\end{align*}
$
<br>
where
<br>
$
C\left(q, \theta\right)
=
\begin{bmatrix}
    M_t & -ml \cos\theta \\
    -ml \cos\theta & J_t\\
\end{bmatrix}
$,
$
D\left(\theta\right) 
=
\begin{bmatrix}
    c & 0 \\
    0 & \gamma -mgl \cos{\theta}\\
\end{bmatrix}
$.

Define the states of the discrete-time system at $t_k$ as $x_t = \left[q(t_k), \theta(t_k), \dot{q}(t_k), \dot{\theta}(t_k) \right]^\top$ and the control inputs $u_k= \left[F_k, 0, 0, 0 \right]^\top$.
<br>
Then, by using the linearlized equation derived above, the dynamics of $x_t$ is linearlzed as,
<br>
$
x_{k+1} \approx A_k x_k + B_k u_k
$,
<br>
where
<br>
$
A_k 
=
\begin{bmatrix}
I & \Delta t I \\
I & -\Delta t C_k^{-1} D_k \\
\end{bmatrix} 
$,
$
B_k
=
\begin{bmatrix}
O & O \\
\Delta t C_k^{-1}
\begin{bmatrix}
1 & 0 \\
0 & 0 \\
\end{bmatrix}
& O\\
\end{bmatrix}
$.
<br>
Here, $I$ and $O$ are the unit, and zero matrices, respectively, and $\Delta t$ is the time step.
<br>

We can also apply the below quadratic cost. Here $Q_k$, $R_k$ and $P_N$ are positive semi-definite. 
<br>
$
c(x_k, u_k) 
=
\begin{cases}
{1\over2} x_k^\top Q_k x_k + {1\over2} u_k^\top R_k u_k  & k<N \text{(stage-wise cost)}\\
{1\over2} x_k^\top P_N x_k  & k=N \text{(terminal cost)}
\end{cases}
$.

With the above formulations, we can contorl the cart-pole system. The code are presented below.

```python
import numpy as np
import scipy as sp
import cvxpy as cp
import matplotlib.pyplot as plt

# Parameters
DYNAMIC_PARAMETERS = {
    'Gravity': -9.8,
    'Cart mass': 10,
    'Cart inertia': 100,
    'Pendulum mass': 80,
    'Pole length': 1.0,
    'Viscousity': 0.1,
    'Damping': 0.01,
    'Time step': 0.05
}

N = 500
```

In this simulation, we set up a consistent time step $\Delta t = 0.05 \text{sec}$, and used a finite time horizon $N=500$ (in simulation time, $25.0 \text{sec}$). For the quadratic cost, we used positive definite diagonal matrices. 

```python
# Define the cart-pole dyanmics
class CartPole(object):

    def __init__(self, parameters):

        self._dt = parameters['Time step']
        self._c = parameters['Viscousity']
        self._r = parameters['Damping']
        self._l = parameters['Pole length']
        self._m = parameters['Pendulum mass']
        self._g = parameters['Gravity']
        self._M_t = parameters['Cart mass'] + parameters['Pendulum mass']
        self._J_t = parameters['Cart inertia'] + (self._m * self._l**2)


    def reset(self, init_state):
        # state = [Cart pos, Pendulum ang, Cart vel, Pendulum angvl]

        self._state = np.array(init_state)


    def _get_mat_C(self, state):

        theta = state[1]
        mat_C = np.array([[self._M_t, - self._m * self._l * np.cos(theta)],
                          [- self._m * self._l * np.cos(theta), self._J_t]])
        
        return mat_C


    def _get_mat_D(self, state):

        theta = state[1]
        mat_D = np.array([[self._c, 0],
                          [0, self._r - self._m * self._g * self._l * np.cos(theta)]])
        
        return mat_D


    def linearize(self, state):

        mat_C = self._get_mat_C(state)
        mat_D = self._get_mat_D(state)
        mat_inv_C = np.linalg.inv(mat_C)

        mat_A = np.zeros((4, 4))
        mat_A[:2, :2] = np.eye(2)
        mat_A[:2, 2:] = self._dt * np.eye(2)
        mat_A[2:, :2] = np.eye(2)
        mat_A[2:, 2:] = -self._dt * mat_inv_C @ mat_D

        mat_B = np.zeros((4, 4))
        mat_B[2:, :2] = self._dt * mat_inv_C @ np.array([[1, 0], [0, 0]])
                
        return mat_A, mat_B


    def step(self, u):

        mat_C = self._get_mat_C(self._state)
        mat_inv_C = np.linalg.inv(mat_C)

        dq = self._state[2]
        dtheta = self._state[3]
        theta = self._state[1]

        vec_nonlin = - self._m * self._l * np.sin(theta) * np.array([dtheta**2, - self._g ])
        vec_lin = - np.array([self._c * dq, self._r * dtheta])
        vec_ctrl = np.array([u[0],0])

        self._state[:2] += self._dt * self._state[2:]
        self._state[2:] += self._dt * mat_inv_C @ (vec_ctrl + vec_lin + vec_nonlin)
        
        return np.copy(self._state)

    
    @property
    def state(self):

        return np.copy(self._state)
```

In `CartPole`, we define the cart-pole system. Every simulation, `reset` initiate the system with the given initial states. For computing iLQR, `linearize` computes the linearized dynamic matrices. For evaluation, `step` computes the forward dynamics. 

```python
def lqr_experiment(env, init_state, target_state):

    print('Computing LQR...')
    
    U = cp.Variable((N,  4))
    X = cp.Variable((N+1,4))
    
    const = []
    const.append(X[0,:] == init_state)
        
    A, B = env.linearize(target_state)
        
    for i in range(N):
        const.append(X[i+1,:] == A @ X[i,:] + B @ U[i,:] ) 

    cost = 0.5e+1*cp.sum((X[:,0]-target_state[0])**2) + \
           1.0e+5*cp.sum((X[:,1]-target_state[1])**2) + \
           1.0e+0*cp.sum((X[:,2]-target_state[2])**2) + \
           4.0e+3*cp.sum((X[:,3]-target_state[3])**2) + \
           1.0e-1*cp.sum((U[:,:])**2)
        
    prob = cp.Problem(cp.Minimize(cost), const)
    prob.solve()

    controls = U.value

    states =  np.zeros((N+1,4))

    env.reset(init_state)
    states[0,:] = env.state
    for i in range(N):
        env.step(controls[i])
        states[i+1,:] = env.state
    
    return {'time': env._dt * np.arange(N), 'states':states[1:], 'controls':controls}
```

In `lqr_experiment` function, we solve the LQR of the discrete-time system linearlized at $\theta=0$, as
<br>
$
x_{k+1} = A x_k + B u_k
$,
<br>
$
Cost = \sum_{k=1}^{N-1} \left(x_k^\top Q x_k + u_k^\top R u_k\right) + x_N^\top Q x_N
$.
<br>
$A$, $B$, $Q$, $R$ are given as the inputs of the function. Here, the linearized model does not rely on the state of $q$.

```python
def ilqr_experiment(env, init_state, target_state):

    print('Computing iLQR...')

    states =  np.zeros((N+1,4))
    controls =  np.zeros((N,4))

    for atempt in range(200):

        U = cp.Variable((N,  4))
        X = cp.Variable((N+1,4))
        
        const = []
        const.append(X[0,:] == np.zeros((4)))
    
        env.reset(init_state)
        states[0,:] = env.state
        A, B = env.linearize(target_state)
        
        for i in range(N):
            env.step(controls[i])
            states[i+1,:] = env.state
            const.append( X[i+1,:] == A @ X[i,:] + B @ U[i,:] ) 

        X_p = X + states
        U_p = U + controls
        
        cost = 0.5e+1*cp.sum((X_p[:,0]-target_state[0])**2) + \
               1.0e+5*cp.sum((X_p[:,1]-target_state[1])**2) + \
               1.0e+0*cp.sum((X_p[:,2]-target_state[2])**2) + \
               4.0e+3*cp.sum((X_p[:,3]-target_state[3])**2) + \
               1.0e-1*cp.sum((U_p[:,:])**2)


        prob = cp.Problem(cp.Minimize(cost), const)
        prob.solve()
        
        d_controls = np.array(U.value)
        controls += d_controls

        if atempt % 10 == 0:
            print('atempt: {} \t control difference: {}'.format(atempt, np.sum(np.sqrt(d_controls**2))/N))

    env.reset(init_state)
    states[0,:] = env.state
    for i in range(N):
        env.step(controls[i])
        states[i+1,:] = env.state        

    return {'time': env._dt * np.arange(N), 'states':states[1:], 'controls':controls}
```

In `ilqr_experiment` function, we solve the iLQR of the discrete-time system, and the computation of the iLQR is iterated at most 200 times.

For each controller, we used the same quadratic cost. Due to the controllability of the system, it is difficult to fully control both $q$ and $\theta$. The main goal of this system is to balance the pole. Therefore, we put higher weights for the terms $\theta$, $\dot\theta$ prioritize the balancing of the pole.   

```python
env = CartPole(DYNAMIC_PARAMETERS)

# Initial state: theta = 10 [deg]
print('Initial state: theta = 10 [deg]')
init_state = np.array([0., 10/180.*np.pi, 0., 0.])
target_state = np.array([0., 0.0, 0., 0.])
result_10 = {}
result_10['LQR'] = lqr_experiment(env, init_state, target_state)
result_10['iLQR'] = ilqr_experiment(env, init_state, target_state)

# Initial state: theta = 30 [deg]
print('Initial state: theta = 30 [deg]')
init_state = np.array([0., 30/180.*np.pi, 0., 0.])
target_state = np.array([0., 0.0, 0., 0.])
result_30 = {}
result_30['LQR'] = lqr_experiment(env, init_state, target_state)
result_30['iLQR'] = ilqr_experiment(env, init_state, target_state)
```

```python
def plot(result):
    
    fig, axs = plt.subplots(5)

    for key in ['LQR', 'iLQR']:
        axs[0].plot(result[key]['time'], result[key]['states'][:,0], label=key)
        axs[1].plot(result[key]['time'], result[key]['states'][:,1], label=key)
        axs[2].plot(result[key]['time'], result[key]['states'][:,2], label=key)
        axs[3].plot(result[key]['time'], result[key]['states'][:,3], label=key)
        axs[4].plot(result[key]['time'], result[key]['controls'][:,0], label=key)
        
    axs[0].set_ylabel("q [m]")
    axs[1].set_ylabel("theta [deg]")
    axs[2].set_ylabel("dq [m/sec]")
    axs[3].set_ylabel("dtheta [deg/sec]")
    axs[4].set_ylabel("Control [Nm]")
    axs[4].set_xlabel("Time [sec]")

    axs[0].set_ylim(-5, 5)
    axs[1].set_ylim(-1, 1)
    axs[2].set_ylim(-5, 5)
    axs[3].set_ylim(-1, 1)
    axs[4].set_ylim(-100, 100)

    axs[4].legend(loc="lower right")

    fig.set_figwidth(8)
    fig.set_figheight(8)

    plt.show()
```

The plots of the intial states $\theta=\text{10}^{\circ}\$ are presented below.
