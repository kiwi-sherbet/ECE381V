## Question 3

The differential equation of the dynamics are given as
<br>
$
\begin{bmatrix} 
    F \\ 0\\
  \end{bmatrix} = 
  \begin{bmatrix}
    M_t & -ml \cos\theta \\
    -ml \cos\theta & J_t\\
  \end{bmatrix}
  \begin{bmatrix}
    \ddot{q} \\
    \ddot{\theta}\\
  \end{bmatrix}
  +
  \begin{bmatrix}
    c & 0 \\
    0 & \gamma\\
  \end{bmatrix}
  \begin{bmatrix}
    \dot{q} \\
    \dot{\theta}\\
  \end{bmatrix}
  +
  \begin{bmatrix}
    ml\dot{\theta}^2\sin{\theta} \\
    -mgl \sin{\theta}\\
  \end{bmatrix}
$.
<br>

We can linearlize the equation as,
<br>
$
\begin{align*}
\begin{bmatrix} 
    F \\ 0\\
\end{bmatrix} 
&\approx 
\begin{bmatrix}
    M_t & -ml \cos\theta \\
    -ml \cos\theta & J_t\\
\end{bmatrix}
\begin{bmatrix}
    \ddot{q} \\
    \ddot{\theta}\\
\end{bmatrix}
+
\begin{bmatrix}
    c & 0 \\
    0 & \gamma\\
\end{bmatrix}
\begin{bmatrix}
    \dot{q} \\
    \dot{\theta}\\
\end{bmatrix}
+
\dot{\theta}
\begin{bmatrix}
    0 \\
    -mgl \cos{\theta}\\
\end{bmatrix}
\\
&= 
C\left(q, \theta\right) 
\begin{bmatrix}
    \ddot{q} \\
    \ddot{\theta}\\
\end{bmatrix}
+
D\left(\theta\right) 
\begin{bmatrix}
    \dot{q} \\
    \dot{\theta}\\
\end{bmatrix}
\end{align*}
$
<br>
where
<br>
$
C\left(q, \theta\right)
=
\begin{bmatrix}
    M_t & -ml \cos\theta \\
    -ml \cos\theta & J_t\\
\end{bmatrix}
$,
$
D\left(\theta\right) 
=
\begin{bmatrix}
    c & 0 \\
    0 & \gamma -mgl \cos{\theta}\\
\end{bmatrix}
$.

Define the states of the discrete-time system at $t_k$ as $x_t = \left[q(t_k), \theta(t_k), \dot{q}(t_k), \dot{\theta}(t_k) \right]^\top$ and the control inputs $u_k= \left[F_k, 0, 0, 0 \right]^\top$.
<br>
Then, by using the linearlized equation derived above, the dynamics of $x_t$ is linearlzed as,
<br>
$
x_{k+1} \approx A_k x_k + B_k u_k
$,
<br>
where
<br>
$
A_k 
=
\begin{bmatrix}
I & \Delta t I \\
I & -\Delta t C_k^{-1} D_k \\
\end{bmatrix} 
$,
$
B_k
=
\begin{bmatrix}
O & O \\
\Delta t C_k^{-1}
\begin{bmatrix}
1 & 0 \\
0 & 0 \\
\end{bmatrix}
& O\\
\end{bmatrix}
$.
<br>
Here, $I$ and $O$ are the unit, and zero matrices, respectively, and $\Delta t$ is the time step.
<br>

We can also apply the below quadratic cost. Here $Q_k$, $R_k$ and $P_N$ are positive semi-definite. 
<br>
$
c(x_k, u_k) 
=
\begin{cases}
{1\over2} x_k^\top Q_k x_k + {1\over2} u_k^\top R_k u_k  & k<N \text{(stage-wise cost)}\\
{1\over2} x_k^\top P_N x_k  & k=N \text{(terminal cost)}
\end{cases}
$.

With the above formulations, we can contorl the cart-pole system. The code are presented below.

```python
import numpy as np
import scipy as sp
import cvxpy as cp
import matplotlib.pyplot as plt

# Parameters
DYNAMIC_PARAMETERS = {
    'Gravity': -9.8,
    'Cart mass': 10,
    'Cart inertia': 100,
    'Pendulum mass': 80,
    'Pole length': 1.0,
    'Viscousity': 0.1,
    'Damping': 0.01,
    'Time step': 0.05
}

N = 500
```

In this simulation, we set up a consistent time step $\Delta t = 0.05 \text{sec}$, and used a finite time horizon $N=500$ (in simulation time, $25.0 \text{sec}$). For the quadratic cost, we used positive definite diagonal matrices. 

```python
# Define the cart-pole dyanmics
class CartPole(object):

    def __init__(self, parameters):

        self._dt = parameters['Time step']
        self._c = parameters['Viscousity']
        self._r = parameters['Damping']
        self._l = parameters['Pole length']
        self._m = parameters['Pendulum mass']
        self._g = parameters['Gravity']
        self._M_t = parameters['Cart mass'] + parameters['Pendulum mass']
        self._J_t = parameters['Cart inertia'] + (self._m * self._l**2)


    def reset(self, init_state):
        # state = [Cart pos, Pendulum ang, Cart vel, Pendulum angvl]

        self._state = np.array(init_state)


    def _get_mat_C(self, state):

        theta = state[1]
        mat_C = np.array([[self._M_t, - self._m * self._l * np.cos(theta)],
                          [- self._m * self._l * np.cos(theta), self._J_t]])
        
        return mat_C


    def _get_mat_D(self, state):

        theta = state[1]
        mat_D = np.array([[self._c, 0],
                          [0, self._r - self._m * self._g * self._l * np.cos(theta)]])
        
        return mat_D


    def linearize(self, state):

        mat_C = self._get_mat_C(state)
        mat_D = self._get_mat_D(state)
        mat_inv_C = np.linalg.inv(mat_C)

        mat_A = np.zeros((4, 4))
        mat_A[:2, :2] = np.eye(2)
        mat_A[:2, 2:] = self._dt * np.eye(2)
        mat_A[2:, :2] = np.eye(2)
        mat_A[2:, 2:] = -self._dt * mat_inv_C @ mat_D

        mat_B = np.zeros((4, 4))
        mat_B[2:, :2] = self._dt * mat_inv_C @ np.array([[1, 0], [0, 0]])
                
        return mat_A, mat_B


    def step(self, u):

        mat_C = self._get_mat_C(self._state)
        mat_inv_C = np.linalg.inv(mat_C)

        dq = self._state[2]
        dtheta = self._state[3]
        theta = self._state[1]

        vec_nonlin = - self._m * self._l * np.sin(theta) * np.array([dtheta**2, - self._g ])
        vec_lin = - np.array([self._c * dq, self._r * dtheta])
        vec_ctrl = np.array([u[0],0])

        self._state[:2] += self._dt * self._state[2:]
        self._state[2:] += self._dt * mat_inv_C @ (vec_ctrl + vec_lin + vec_nonlin)
        
        return np.copy(self._state)

    
    @property
    def state(self):

        return np.copy(self._state)
```

In `CartPole`, we define the cart-pole system. Every simulation, `reset` initiate the system with the given initial states. For evaluation, `step` computes the forward dynamics. 

```python
def qp_experiment(env, init_state, target_state):

    print('Computing QP...')
    
    Z = cp.Variable((2*N+1) * 4)

#     W = np.zeros(((2*N+1) * 4, (2*N+1) * 4))
    C = np.zeros(((N+1) * 4, (2*N+1) * 4))
    d = np.zeros(((N+1)*4))

    Q_k = np.zeros((4,4))
    Q_k[0, 0] = 0.5e+1
    Q_k[1, 1] = 1.0e+5
    Q_k[2, 2] = 4.0e+3
    Q_k[3, 3] = 1.0e-1

    R_k = 1.0e-1*np.eye(4)
    
    A, B = env.linearize(target_state)

    for idx in range(N):
#         W[4*2*idx:4*(2*idx+1), 4*2*idx:4*(2*idx+1)] = np.copy(Q_k)
#         W[4*(2*idx+1):4*2*(idx+1), 4*(2*idx+1):4*2*(idx+1)] = np.copy(R_k)

        C[4*idx:    4*(idx+1), 4*2*idx:    4*(2*idx+1)] = -np.eye(4)
        C[4*(idx+1):4*(idx+2), 4*2*idx:    4*(2*idx+1)] = np.copy(A)
        C[4*(idx+1):4*(idx+2), 4*(2*idx+1):4*2*(idx+1)] = np.copy(B)

    C[4*N:4*(N+1), 4*2*N:4*(2*N+1)] = - np.eye(4)
#     W[4*2*N:4*(2*N+1), 4*2*N:4*(2*N+1)] = np.copy(Q_k)
        
        
    d[:4] = np.copy(init_state)
    
    const = []        
    const.append(0 == C @ Z + d ) 
#     cost = Z.T @ W @ Z
#     ## Due to issues at CVXPY, we explicitly define the cost function rather than using W. 
#     ## The computation results of the cost function are the same.
    cost = 0.5e+1*cp.sum((Z[0::8]-target_state[0])**2) + \
           1.0e+5*cp.sum((Z[1::8]-target_state[1])**2) + \
           1.0e+0*cp.sum((Z[2::8]-target_state[2])**2) + \
           4.0e+3*cp.sum((Z[3::8]-target_state[3])**2) + \
           1.0e-1*cp.sum((Z[4::8])**2) + \
           1.0e-1*cp.sum((Z[5::8])**2) + \
           1.0e-1*cp.sum((Z[6::8])**2) + \
           1.0e-1*cp.sum((Z[7::8])**2)
    
    prob = cp.Problem(cp.Minimize(cost), const)
    prob.solve()

    solutions = Z.value
    solutions = np.array(solutions).reshape((2*N+1, 4))
    controls = solutions[1::2,:]
    
    states =  np.zeros((N+1,4))

    env.reset(init_state)
    states[0,:] = env.state
    for i in range(N):
        env.step(controls[i])
        states[i+1,:] = env.state
    
    return {'time': env._dt * np.arange(N), 'states':states[1:], 'controls':controls}
```

In `qp_experiment` function, we solve the LQR of the discrete-time system linearlized at $\theta=0$, as the formulation of the quadratic program at Q2.
<br>
$
W Z + d = 0
$,
<br>
$
Cost = Z^\top W Z
$.


```python
def lqr_experiment(env, init_state, target_state):

    print('Computing LQR...')
    
    U = cp.Variable((N,  4))
    X = cp.Variable((N+1,4))
    
    const = []
    const.append(X[0,:] == init_state)
        
    A, B = env.linearize(target_state)
        
    for i in range(N):
        const.append(X[i+1,:] == A @ X[i,:] + B @ U[i,:] ) 

    cost = 0.5e+1*cp.sum((X[:,0]-target_state[0])**2) + \
           1.0e+5*cp.sum((X[:,1]-target_state[1])**2) + \
           1.0e+0*cp.sum((X[:,2]-target_state[2])**2) + \
           4.0e+3*cp.sum((X[:,3]-target_state[3])**2) + \
           1.0e-1*cp.sum((U[:,:])**2)
        
    prob = cp.Problem(cp.Minimize(cost), const)
    prob.solve()

    controls = U.value

    states =  np.zeros((N+1,4))

    env.reset(init_state)
    states[0,:] = env.state
    for i in range(N):
        env.step(controls[i])
        states[i+1,:] = env.state
    
    return {'time': env._dt * np.arange(N), 'states':states[1:], 'controls':controls}
```

In `lqr_experiment` function, we solve the LQR of the discrete-time system linearlized at $\theta=0$, using dynamic programming, as
<br>
$
x_{k+1} = A x_k + B u_k
$,
<br>
$
Cost = \sum_{k=1}^{N-1} \left(x_k^\top Q x_k + u_k^\top R u_k\right) + x_N^\top Q x_N
$.
<br>
$A$, $B$, $Q$, $R$ are given as the inputs of the function. Here, the linearized model does not rely on the state of $q$.

For each method, we used the same quadratic cost. Due to the controllability of the system, it is difficult to fully control both $q$ and $\theta$. The main goal of this system is to balance the pole. Therefore, we put higher weights for the terms $\theta$, $\dot\theta$ prioritize the balancing of the pole.   

```python
env = CartPole(DYNAMIC_PARAMETERS)

# Initial state: theta = 10 [deg]
print('Initial state: theta = 10 [deg]')
init_state = np.array([0., 10/180.*np.pi, 0., 0.])
target_state = np.array([0., 0.0, 0., 0.])
result_10 = {}
result_10['QP'] = qp_experiment(env, init_state, target_state)
result_10['DP'] = lqr_experiment(env, init_state, target_state)

# Initial state: theta = 30 [deg]
print('Initial state: theta = 30 [deg]')
init_state = np.array([0., 30/180.*np.pi, 0., 0.])
target_state = np.array([0., 0.0, 0., 0.])
result_30 = {}
result_30['QP'] = qp_experiment(env, init_state, target_state)
result_30['DP'] = lqr_experiment(env, init_state, target_state)
```

```python
def plot(result):
    
    fig, axs = plt.subplots(5)

    for key in ['DP', 'QP']:
        axs[0].plot(result[key]['time'], result[key]['states'][:,0], label=key)
        axs[1].plot(result[key]['time'], result[key]['states'][:,1], label=key)
        axs[2].plot(result[key]['time'], result[key]['states'][:,2], label=key)
        axs[3].plot(result[key]['time'], result[key]['states'][:,3], label=key)
        axs[4].plot(result[key]['time'], result[key]['controls'][:,0], label=key)
        
    axs[0].set_ylabel("q [m]")
    axs[1].set_ylabel("theta [deg]")
    axs[2].set_ylabel("dq [m/sec]")
    axs[3].set_ylabel("dtheta [deg/sec]")
    axs[4].set_ylabel("Control [Nm]")
    axs[4].set_xlabel("Time [sec]")
    axs[4].legend(loc="lower right")

    fig.set_figwidth(8)
    fig.set_figheight(8)

    plt.show()
```

The plots of the intial states $\theta=\text{10}^{\circ}\$ are presented below.

```python
plot(result_10)
```

The plots of the intial states $\theta=\text{30}^{\circ}\$ are presented below.

```python
plot(result_30)
```

We can find that the optimal control actions obtained using the two methods match each other.
